```python
import pandas as pd
```

```python
df = pd.read_csv("corporate_rating.csv")
df.head()
```

```python
df.info()
```

```python
df['Sector'].unique()
```

```python
sector_mapping = {
    'Consumer Durables': 'Project Developers',
    'Energy': 'EPC Providers, O&M Providers, Utility Companies, Project Developers',
    'Capital Goods': 'EPC Providers, O&M Providers',
    'Consumer Non-Durables': 'Project Developers',
    'Public Utilities': 'Utility Companies, O&M Providers',
    'Technology': 'O&M Providers, Project Developers',
    'Health Care': 'Project Developers',
    'Consumer Services': 'Project Developers',
    'Miscellaneous': 'Varies',
    'Basic Industries': 'EPC Providers, O&M Providers',
    'Finance': 'Indirect Role (e.g., Funding via Project Developers)',
    'Transportation': 'EPC Providers, O&M Providers'
}

# Apply mapping to the dataframe
df['Mapped Category'] = df['Sector'].map(sector_mapping)
```

```python
rating_dict = {'AAA':'Lowest Risk', 
               'AA':'Low Risk',
               'A':'Low Risk',
               'BBB':'Medium Risk', 
               'BB':'High Risk',
               'B':'High Risk',
               'CCC':'Highest Risk', 
               'CC':'Highest Risk',
               'C':'Highest Risk',
               'D':'In Default'}

df.Rating = df.Rating.map(rating_dict)
```

```python
# donwload csv here with rating on differnt years
df.to_csv('Rating_with_history.csv', index=False)
```

```python
rating_to_score = {
    'Lowest Risk' : 5 , 'Low Risk' : 4 , 'Medium Risk' : 3 , 'High Risk' : 2 , 'Highest Risk':1
}

# Map ratings to numerical scores
df['Rating_Score'] = df['Rating'].map(rating_to_score)

# Calculate the average score for each company
aggregated = df.groupby('Name').agg(
    Average_Score=('Rating_Score', 'mean')  # Calculate the average of numeric scores
).reset_index()

# Optionally map the average score back to categorical ratings
score_to_rating = {v: k for k, v in rating_to_score.items()}
aggregated['Average_Rating'] = aggregated['Average_Score'].apply(lambda x: score_to_rating[round(x)])

# Display results
print(aggregated)
```

```python
aggregated.to_csv("Rating_without_history.csv",index=False)
```
